```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv') 
```

```python
# Display the 2020 sales DataFrame
athletic_2020_df
```

```python
# Display the 2021 sales DataFrame
athletic_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.()
athletic_2020_df.info()
```

```python
# Check the 2021 sales data types.
athletic_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_data_rows_df = pd.concat([athletic_2020_df, athletic_2021_df], axis='rows', join="inner").reset_index(drop=True)
joined_data_rows_df
```

```python
# Check if any values are null.
joined_data_rows_df.isnull().sum()
```

```python
# Check the data type of each column
joined_data_rows_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_data_rows_df['invoice_date'] = pd.to_datetime(joined_data_rows_df['invoice_date'])
joined_data_rows_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_data_rows_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
    
total_products_sold = joined_data_rows_df.groupby(['region', 'state', 'city'])[['units_sold']].sum()
    
# Rename the sum to "Total_Products_Sold".
total_products_sold.rename(columns = {'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
total_products_sold.sort_values(by=['Total_Products_Sold'], ascending=False).head()

```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold = pd.pivot_table(joined_data_rows_df, index=['region', 'state', 'city'], values='units_sold', aggfunc='sum').round(1)

# Rename the "units_sold" column to "Total_Products_Sold"
total_products_sold.rename(columns={'units_sold':'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
total_products_sold.sort_values(by=['Total_Products_Sold'], ascending=False).head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_amount = joined_data_rows_df.groupby(['region', 'state', 'city'])[['total_sales']].sum()

# Rename the "total_sales" column to "Total Sales"
total_sales_amount.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Show the top 5 results.
total_sales_amount.sort_values(by=['Total_Sales'], ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_amount = pd.pivot_table(joined_data_rows_df, index=['region', 'state', 'city'], values='total_sales', aggfunc='sum').round(1)

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_amount.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Show the top 5 results.
total_sales_amount.sort_values(by=['Total_Sales'], ascending=False).head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_by_retailer = joined_data_rows_df.groupby(['retailer', 'region', 'state', 'city'])[['total_sales']].sum()

# Rename the "total_sales" column to "Total Sales"
total_sales_by_retailer.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Show the top 5 results.
total_sales_by_retailer.sort_values(by=['Total_Sales'], ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_by_retailer = pd.pivot_table(joined_data_rows_df, index=['retailer', 'region', 'state', 'city'], values='total_sales', aggfunc='sum').round(1)

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_by_retailer.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Show the top 5 results.
total_sales_by_retailer.sort_values(by=['Total_Sales'], ascending=False).head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear_df = joined_data_rows_df[joined_data_rows_df['product'].str.contains("Women's Athletic Footwear")]
womens_athletic_footwear_df


```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
total_womens_athletic_footwear = womens_athletic_footwear_df.groupby(['retailer', 'region', 'state', 'city'])[['units_sold']].sum()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_womens_athletic_footwear.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
total_womens_athletic_footwear.sort_values(by=['Womens_Footwear_Units_Sold'], ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
total_womens_athletic_footwear = pd.pivot_table(womens_athletic_footwear_df, index=['retailer', 'region', 'state', 'city'], values='units_sold', aggfunc='sum').round(1)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_womens_athletic_footwear.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
total_womens_athletic_footwear.sort_values(by=['Womens_Footwear_Units_Sold'], ascending=False).head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_athletic_footwear_sales_by_date = pd.pivot_table(womens_athletic_footwear_df, index=['invoice_date'], values='total_sales', aggfunc='sum').round(1)

# Optional: Rename the "total_sales" column to "Total Sales"
womens_athletic_footwear_sales_by_date.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Show the table.
womens_athletic_footwear_sales_by_date.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_bins_df = womens_athletic_footwear_df.set_index('invoice_date') 
daily_bins_df = daily_bins_df.resample('D').sum()

daily_bins_df.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Sort the resampled pivot table in descending order on "Total Sales".
daily_bins_df.sort_values(by=['Total_Sales'], ascending=False).head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_bins_df = womens_athletic_footwear_df.set_index('invoice_date') 
weekly_bins_df = weekly_bins_df.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
weekly_bins_df.sort_values(by=['total_sales'], ascending=False).head(10)
```
